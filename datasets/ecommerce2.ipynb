```python
"""All available columns:
['order_id', 'order_num', 'date', 'email', 'status', 'currency',
       'subtotal', 'taxes', 'shipping', 'shipping_taxes', 'total', 'discount',
       'promo_code', 'payment_mode', 'invoice_num', 'receipt_num',
       'has_invoice', 'shipping_method', 'article_quantity', 'article_title',
       'article_options_variants', 'article_price', 'article_sku', 'last_name',
       'first_name', 'phone', 'company', 'vat_number',
       'shipping_address_last_name', 'shipping_address_first_name',
       'shipping_address', 'billing_address', 'shipping_tracking_num',
       'customer_note', 'internal_note', 'origin', 'tax_amount_at_0.0',
       'taxable_amount_at_0.0', 'invoice_url', 'order_summary_url']"""

drop_columns = ['email', 'status', 'currency','subtotal','taxes','shipping', 'shipping_taxes', 'discount','payment_mode','invoice_num','receipt_num','has_invoice','article_options_variants','article_price','last_name','first_name','phone', 'company', 'vat_number',
       'shipping_address_last_name', 'shipping_address_first_name',
       'shipping_address', 'billing_address', 'shipping_tracking_num',
       'customer_note', 'internal_note', 'tax_amount_at_0.0',
       'taxable_amount_at_0.0', 'invoice_url', 'order_summary_url']
df.drop(columns=drop_columns, axis=1, inplace=True)


    #Converting dtypes float into int
float_cols=df.select_dtypes(include=['float64']).columns
df[float_cols] = df[float_cols].fillna(0).astype('int64')
    
    #Converting 'total' column into proper format
df['total'] = (
df['total'].fillna(method='ffill')
.str.replace('.','', regex=False)
.astype(int)
//100_000
)

    #Converting dtypes date into datetime
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y', errors='coerce')
df['date'] = df['date'].ffill()

    #Dealing null values
df['promo_code'] = df['promo_code'].fillna('None')
df[['shipping_method','origin']] = df[['shipping_method','origin']].ffill()
df[['order_id','order_num']]=df[['order_id','order_num']].astype('str').replace('0', pd.NA).ffill()
```
